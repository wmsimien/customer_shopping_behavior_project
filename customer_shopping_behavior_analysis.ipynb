### Customer Shopping Behavior Analysis

```python
# import libraries
import pandas as pd
```

```python
# import data from csv file; create data frame
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
df.isnull().sum() # check null values; mean or medium; mean is effective by outliers; median(middle) is robost to outliers
```

```python
# impute the review rate per category w/ median for that combination
# df.groupby('Category').agg({'Review Rating': 'median'})
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# rename columns using snake case
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group to group customers based on age
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create column purchage_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 30,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```

```python
df[['discount_applied', 'promo_code_used']]
# are they the same; if so, don't need both columns
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# drop column promo_code_used
df = df.drop('promo_code_used', axis=1)
df.columns
```

```python
# data analayed, cleaned, transformed; let's move to SQL

# import libraries from psycog2 and sqlalchemy to connect to PostgreSQL 18
import psycopg2
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
# load .env; loads credentials into memeory 
load_dotenv(override=True)
```

```python
# get credentials
host=os.getenv('DB_HOST')
username=os.getenv('DB_USER')
password=os.getenv('DB_PASSWORD')
database=os.getenv('DB_NAME')
port=os.getenv('DB_PORT')
```

```python
# create connection to PostgreSQL db
connection_string = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(connection_string)

# load data frame into db
table_name = 'customer'
df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
